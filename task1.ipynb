```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('/content/sample_data/population_by_age_group.csv')
```

```python
# Calculate global totals for each age group
global_totals = df[['65+', '25-64 years', '15-24 years', '5-14 years', '0-4 years']].sum()
```

```python
# Convert to percentages for better visualization
global_percentages = (global_totals / global_totals.sum()) * 100
```

```python
# Create a histogram/bar chart
plt.figure(figsize=(12, 7))
plt.bar(global_percentages.index, global_percentages.values,
        color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

plt.title('Global Population Distribution by Age Group', fontsize=16, pad=20)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Percentage of Global Population', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

```

```python
# Adjust layout to prevent label cutoff
plt.tight_layout()
plt.show()
```
